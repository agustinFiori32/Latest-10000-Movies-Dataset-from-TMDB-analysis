```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
from matplotlib import cm
from matplotlib import colors
```

```python
df = pd.read_csv('Movies_dataset.csv')
```

```python
df.info()
```

```python
df.columns
```

```python
top_to_analysis = int(input('Ingrese la cantidad de peliculas a analizar: '))
```

```python
df_top_popularity = df.sort_values(by='popularity', ascending=False)            #Create a new df with the 15 most popular movies order by our qualification
df_top_popularity = df_top_popularity[df_top_popularity['vote_count'] > 1900].head(top_to_analysis)
df_top_popularity = df_top_popularity.sort_values(by='vote_average', ascending=True)
df_top_popularity
```

```python
df_top_popularity_for_popularity = df_top_popularity.sort_values(by='popularity', ascending=True)
df_top_popularity_for_popularity
```

```python
plt.figure()
plt.barh(df_top_popularity_for_popularity['title'],df_top_popularity_for_popularity['popularity'],align='center',height=0.7,color='#779ECB')
plt.title(f'Top {top_to_analysis} most popular movies')
plt.show()
```

```python
plt.figure()
plt.barh(df_top_popularity['title'],df_top_popularity['vote_average'],align='center',height=0.7,color='#779ECB')
plt.title(f'Top {top_to_analysis} most popular movies by vote average')
plt.show()
```

```python
df_top_votes_for_average = df.sort_values(by='vote_average', ascending=False)
df_top_votes_for_average = df_top_votes_for_average[df_top_votes_for_average['vote_count'] > 1900].head(top_to_analysis)
df_top_votes_for_average = df_top_votes_for_average.sort_values(by='vote_average', ascending=True)
df_top_votes_for_average
```

```python
plt.figure()
plt.barh(df_top_votes_for_average['title'],df_top_votes_for_average['vote_average'],align='center',height=0.7,color='#779ECB')
plt.title(f'Top {top_to_analysis} best movies')
plt.show()
```

```python
df_top_votes = df.sort_values(by='vote_average', ascending=False)           #Create a similar df with the top rated movies order by popularity
df_top_votes = df_top_votes[df_top_votes['vote_count'] > 1900].head(top_to_analysis)
df_top_votes = df_top_votes.sort_values(by='popularity', ascending=True)
df_top_votes
```

```python
plt.figure()
plt.barh(df_top_votes['title'],df_top_votes['popularity'],align='center',height=0.7,color='#779ECB')
plt.title(f'Top {top_to_analysis} best movies by popularity')
plt.show()
```

```python
languages_count_popular = df_top_popularity['original_language'].value_counts()        #Observe the language trend of the most popular movies
languages_count_popular  #English dominate
```

```python
languages_count_bests = df_top_votes['original_language'].value_counts()        #The same for the top rating movies
languages_count_bests   #More than English
```

```python
#We modify the index for better understanding
traductions = {
    'en' : 'English',
    'it' : 'Italian',
    'ja' : 'Japanese',
    'ko' : 'Korean',
    'hi' : 'Hindi'    
}

languages_count_bests_full = languages_count_bests.index.map(traductions)
languages_count_bests.index = languages_count_bests_full
languages_count_bests
```

```python
#We embellish the pie chart ;)
min_value = min(languages_count_bests.values)
adjusted_min_value = min_value * 0.0000000000000000000000000000000001
def custom_norm(x):
    if x <= adjusted_min_value:
        return 0.00000000000000000000000000001
    return (x - adjusted_min_value) / (max(languages_count_bests.values) - adjusted_min_value)
cmap = cm.get_cmap('Blues')
colores = [cmap(custom_norm(val)) for val in languages_count_bests.values]

plt.pie(languages_count_bests.values, labels=languages_count_bests.index, autopct="%0.1f %%", colors= colores)
plt.title(f'Languages of the {top_to_analysis} top rating movies')
plt.show()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce',dayfirst=True)
df_top_popularity['release_date'] = pd.to_datetime(df_top_popularity['release_date'], errors='coerce',dayfirst=True)
df_top_votes['release_date'] = pd.to_datetime(df_top_votes['release_date'], errors='coerce',dayfirst=True)
#I need the release date in correct format for analysis
```

```python
df['release_year'] = df['release_date'].dt.year
df_top_popularity['release_year'] = df_top_popularity['release_date'].dt.year
df_top_votes['release_year'] = df_top_votes['release_date'].dt.year
```

```python
decades = list(range(1900, 2030, 10))       #Create a df with the decades to analysis
decades_df = pd.DataFrame({'decades': decades})
decades_df['count_movies'] = [len(df[(df['release_year'] >= start) & (df['release_year'] < start + 10)]) for start in decades]
decades_df = decades_df.set_index('decades')
decades_df
```

```python
plt.plot(decades_df.index, decades_df.values, marker='o', markersize=5,color='#779ECB')
plt.title('Movies in the dataset for decade')
plt.show()
```

```python
decades_df_popularity = pd.DataFrame({'decades': decades})          #The decades of 2010 and 2020 dominate the popularity
decades_df_popularity['count_movies'] = [len(df_top_popularity[(df_top_popularity['release_year'] >= start) & (df_top_popularity['release_year'] < start + 10)]) for start in decades]
decades_df_popularity = decades_df_popularity.set_index('decades')
decades_df_popularity = decades_df_popularity[decades_df_popularity['count_movies'] != 0]
decades_df_popularity
```

```python
plt.bar(decades_df_popularity.index, decades_df_popularity['count_movies'], width=8, align='center',color='#779ECB')
plt.gca().set_facecolor('#D0D0D0')
plt.title('Top popular movies for decade')
plt.show()
```

```python
decades_df_rating = pd.DataFrame({'decades': decades})          #In rating is not the same
decades_df_rating['count_movies'] = [len(df_top_votes[(df_top_votes['release_year'] >= start) & (df_top_votes['release_year'] < start + 10)]) for start in decades]
decades_df_rating = decades_df_rating.set_index('decades')
decades_df_rating = decades_df_rating[decades_df_rating['count_movies'] != 0]
decades_df_rating
```

```python
plt.bar(decades_df_rating.index, decades_df_rating['count_movies'], width=8, align='center',color='#779ECB')
plt.gca().set_facecolor('#D0D0D0')
plt.title('Top rating movies for decades')
plt.show()
```
